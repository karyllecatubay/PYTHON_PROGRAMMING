## CHALLENGE: CONVERT TEMPERATURES

```python
def convert_far_to_cel(fahrenheit):
    celsius = (fahrenheit - 32) * 5/9
    return celsius
```

```python
temp_f = int(input("Enter a temperature in degrees F:"))
temp_c = convert_far_to_cel(temp_f)
print(f"{temp_f} degrees F = {temp_c:.2f} degrees C")
```

```python
def convert_cel_to_far(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit
```

```python
temp_c = int(input("Enter a temperature in degrees C:"))
temp_f = convert_cel_to_far(temp_c)
print(f"{temp_c} degrees C = {temp_f:.2f} degrees F")
```
