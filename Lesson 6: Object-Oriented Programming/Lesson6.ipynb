### Review Question 1

```python
class Dog:
    def __init__(self, name, age, coat_color):
        self.name = name
        self.age = age
        self.coat_color = coat_color

philo = Dog("Philo", 5, "brown")
print(f"{philo.name}'s coat is {philo.coat_color}.")
```

```python
class Car:
    def __init__(self, color, mileage):
        self.color = color
        self.mileage = mileage

blue_car = Car("blue", 20000)
red_car = Car("red", 30000)

print(f"The {blue_car.color} car has {blue_car.mileage:,} miles.")
print(f"The {red_car.color} car has {red_car.mileage:,} miles.")
```

```python
class Car:
    def __init__(self, color, mileage):
        self.color = color
        self.mileage = mileage

    def drive(self, miles):
        self.mileage += miles

test_car = Car("gray", 0)
test_car.drive(100)
print(test_car.mileage)
```

### Review Question # 2

```python
class Dog:
    species = "Canis familiaris"

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name} is {self.age} years old"

    def speak(self, sound):
        return f"{self.name} says {sound}"

class GoldenRetriever(Dog):
    def speak(self, sound="bark"):
        return super().speak(sound)
```

```python
golden = GoldenRetriever("Buddy", 3)
print(golden.speak())            # Output: Buddy says bark
print(golden.speak("woof"))      # Output: Buddy says woof
```

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

class Square(Rectangle):
    def __init__(self, side_length):
        super().__init__(side_length, side_length)

square = Square(4)
print(square.area())
```

### CHALLENGE: MODEL A FARM

```python
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def eat(self):
        return f"{self.name} is eating."

    def sleep(self):
        return f"{self.name} is sleeping."

class Cow(Animal):
    def moo(self):
        return f"{self.name} says Moo!"

class Chicken(Animal):
    def lay_egg(self):
        return f"{self.name} laid an egg."

class Sheep(Animal):
    def shear(self):
        return f"{self.name} is being sheared."

# Instantiate the animals
bessie = Cow("Bessie", 4)
clucky = Chicken("Clucky", 2)
woolly = Sheep("Woolly", 3)

# Output behaviors and details
print(f"{bessie.name} is a {bessie.age}-year-old cow.")
print(bessie.eat())
print(bessie.moo())

print(f"{clucky.name} is a {clucky.age}-year-old chicken.")
print(clucky.sleep())
print(clucky.lay_egg())

print(f"{woolly.name} is a {woolly.age}-year-old sheep.")
print(woolly.eat())
print(woolly.shear())
```
